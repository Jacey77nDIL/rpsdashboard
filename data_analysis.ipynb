```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ["Rock", "Paper", "Scissors"]
bot_wins = [bot_rock_win_count, bot_paper_win_count, bot_scissors_win_count]  # Wins by the Python bot
computer_wins = [computer_rock_win_count, computer_paper_win_count, computer_scissors_win_count]  # Wins by the computer

x = np.arange(len(categories))  # X-axis positions for bars
width = 0.4  # Width of bars

# Create grouped bars
plt.bar(x - width/2, bot_wins, width=width, label="Bot Wins", color="red")
plt.bar(x + width/2, computer_wins, width=width, label="Computer Wins", color="blue")

# Labels and title
plt.xlabel("Move")
plt.ylabel("Win Count")
plt.title("Win Comparison: Bot vs Computer")
plt.xticks(x, categories)  # Set x-axis labels
plt.yticks(np.arange(0, max(bot_wins + computer_wins) + 10, 5))
# Add legend
plt.legend()

# Show plot
plt.show()
```

```python
# Data
categories = ["Rock", "Paper", "Scissors"]
bot_loss = [bot_rock_loss_count, bot_paper_loss_count, bot_scissors_loss_count]  # loss by the Python bot
computer_loss = [computer_rock_loss_count, computer_paper_loss_count, computer_scissors_loss_count]  # loss by the computer

x = np.arange(len(categories))  # X-axis positions for bars
width = 0.4  # Width of bars

# Create grouped bars
plt.bar(x - width/2, bot_loss, width=width, label="Bot loss", color="red")
plt.bar(x + width/2, computer_loss, width=width, label="Computer loss", color="blue")

# Labels and title
plt.xlabel("Move")
plt.ylabel("loss Count")
plt.title("loss Comparison: Bot vs Computer")
plt.xticks(x, categories)  # Set x-axis labels
plt.yticks(np.arange(0, max(bot_wins + computer_wins) + 10, 10))
# Add legend
plt.legend()

# Show plot
plt.show()
```
